```python
import matplotlib.pyplot as plt

student1 = {'chinese':78, 
            'english':80,
            'math':92,
            'history':75,
            'discover':85}

student2 = {'chinese':69, 
            'english':92,
            'math':74,
            'history':69,
            'discover':93}

subjects = list(student1.keys())
score1 = list(student1.values())

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,score1,'ro--',label='Sudent1')

score2 = list(student2.values())
axes.plot(subjects,score2,'bo-',label='Student2')

axes.set_title('Table Of Students')
axes.set_xlabel('Subject')
axes.set_ylabel('Scores')
axes.grid(axis='y')
axes.legend()

x_locations = axes.get_xticks()
print(x_locations)
for i in x_locations:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+0.5))

plt.show()
```

```python
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
#font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#使用codespace 要安裝中文字型
#install 要加-y
#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plti
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

student1 = {'國文':78, 
            '英文':80,
            '數學':92,
            '歷史':75,
            '地理':85}

student2 = {'國文':69, 
            '英文':92,
            '數學':74,
            '歷史':69,
            '地理':93}

subjects = list(student1.keys())
score1 = list(student1.values())

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,score1,'ro--',label='學生1')

score2 = list(student2.values())
axes.plot(subjects,score2,'bo-',label='學生2')

axes.set_title('學生分數表',fontproperties=font)
axes.set_xlabel('科目',fontproperties=font)
axes.set_ylabel('分數',fontproperties=font)
axes.set_xticklabels(subjects,fontproperties=font)
axes.grid(axis='y')
axes.legend(prop=font)

x_locations = axes.get_xticks()
print(x_locations)
for i in x_locations:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+0.5))

plt.show()
```
